to give bonus points
to the top 25% of students based on their math score

```python
import numpy as np
```

```python
math_scr = df['math score']
```

```python
cutoff_scr = np.percentile(math_scr,75)
```

```python
top25_percnt = df[df['math score'] >= cutoff_scr]
```

```python
bonus_point = 5
top25_percnt['bonus_point'] = bonus_point
top25_percnt.to_csv('students_with_bonus.csv',index=False)
```
